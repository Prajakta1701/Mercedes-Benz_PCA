**Selecting the features and label from the train set and performing the split**

```python
features = MB_Train_data.iloc[:,1:]
target= MB_Train_data.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, train_size=0.80, random_state = 4)

from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler().fit(X_train)
X_train_scaled = scaling.transform(X_train)
test_scaled = scaling.transform(MB_Test_data)
```

### Performing linear regression

```python
from sklearn.linear_model import LinearRegression
lireg = LinearRegression()
lireg.fit(X_train,y_train)
```

```python
y_pred = lireg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred, y_test)
```

Above high value of MSE (1.69) suggests the biasedness or high variance or both in the given dataset.
Therefore, the model is not best fit for the given dataset.

To remove the biasedness in the dataset performing the Principal Component analysis. PCA will help to reduce the dimensionality of the given dataset

### Performing PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 'mle')
pca.fit(X_train)
```

```python
pca.explained_variance_ratio_
```

```python
X_train_transformed = pca.transform(X_train)
X_test_transformed = pca.transform(X_test)
```

```python
X_train_transformed.shape
```

As shown above the PCA has reduced the features from 376 to 255.

### Performing the linear regression on the data derived from the PCA

```python
Lireg2 = LinearRegression()
Lireg2.fit(X_train_transformed, y_train)
```

```python
y_pred2 = Lireg2.predict(X_test_transformed)
y_pred2
```

```python
mean_squared_error(y_pred2, y_test)
```

Even after performing the PCA, the MSE value of the regression is still high 74.04 which makes the model less reliable.

To avoid the problem of biasedness and overfitting using the XGBoost algorithm

### Performing XGBoost algorithm

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
```

```python
grid_values = {'n_estimators': [74,75,76], 'learning_rate': [0.13,0.135,0.14], 'max_depth': [1,2,3]}
grid_xgb = GridSearchCV(xgb.XGBRegressor(), param_grid = grid_values, cv=10, scoring = 'r2')
grid_xgb.fit(X_train_transformed, y_train)
```

```python
print('Mean score matrix: ', grid_xgb.cv_results_['mean_test_score'])
print('Grid best parameter (max. accuracy): ', grid_xgb.best_params_)
print('Grid best score (accuracy): ', grid_xgb.best_score_)
```

The given R2 value suggests the variance of the dependent variable explained by the independent variable. A relatively high R2 value here, makes the model best fit.

```python
predict_xgb = grid_xgb.predict(X_test_transformed)
```

**Saving the final prediction in a CSV file**

```python
Predictions = pd.DataFrame()
Predictions['y'] = pd.Series(predict_xgb)
Predictions.to_csv('Predictions.csv')
Predictions
```
